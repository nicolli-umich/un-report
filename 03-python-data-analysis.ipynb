```python
# only want to look at one variable: aggregate life expectancy data
(
    gapminder
    .agg({'lifeExp': 'mean'})
)
```

```python
# display the descriptives at a custom header
(
    gapminder
    .agg(mean_lifeExp = ('lifeExp', 'mean'))
)
```

```python
# See the quantiles at 0.05 and 0.95. 
(
    gapminder
    .agg({'lifeExp': lambda x: x.quantile([0.05, 0.95])})
)
```

```python
# narrow down our data with query
(
    gapminder
    .query("year == 2007") #R equivalent: dplyr::filter. Always use double quotes
    .agg({'lifeExp': 'mean'})
)
```

```python
# These two blocks are identical, same as dplyr::filter
(
    gapminder
    .query("country == 'United States' and year > 2000") # put the string within single brackets
)

(
    gapminder
    .query("country == 'United States'")
    .query("year > 2000")
)
```

```python
(
    gapminder
    .query("country in ['United States', 'Canada']")
)
```

```python
country_list = ["United States", "Canada"]
```

```python
(
    gapminder
    .query("country in @country_list")
)
```

```python
# define NOT in a list
(
    gapminder
    .query("year == 2007")
    .query("continent == 'Americas'")
    .query("country not in @country_list")
)
```

```python
# Groupby function: R equivalent: dplyr::group_by
(
    gapminder
    .groupby('year')
    .agg({'lifeExp' : ['mean', 'min', 'max']})
)
```

```python
(
    gapminder
    .groupby('continent')
    .agg({'lifeExp' : ['mean', 'min', 'max']})
)
```

```python
# Sorting the data in the order of lifeExp
(
    gapminder
    .query("year == 2007")
    .groupby('continent')
    .agg({'lifeExp': 'mean'})
    .sort_values('lifeExp', ascending = False)
)
```

```python
# .Assign, R equivalent: dplyr::mutate
(
    gapminder
    .assign(gdp = lambda df: df['pop'] * df['gdpPercap'], 
           popInMillions = lambda df: df['pop'] / 1e6)
)
```

```python
# Trick: getting columns from the dataset
gapminder[['country', 'year', 'pop']]
```

### Select and Drop Columns

```python
# Selecting certain columns: Equivalent to dplyr::select
columns_wanted = ['country', 'year', 'pop']
(
    gapminder
    .filter(columns_wanted) 
)
```

```python
# Dropping columns
columns_notwanted = ['year', 'pop']
(
    gapminder
    .drop(columns = columns_notwanted)
)
```

```python
# Selecting for columns that start with c
(
    gapminder
    .filter(regex = '^c')
)
```

```python
# Selecting for columns that end with p
(
    gapminder
    .filter(regex = 'p$')
)
```

### Long to wide format

```python
(
    gapminder
    .pivot(columns = 'year', # the columns need to be the thing that repeats across the observations
          index = ['country', 'continent'],
          values = 'lifeExp')
)
```

```python
(
    gapminder
    .pivot(columns = 'year',
          index = ['country', 'continent'],
          values = 'lifeExp')
    .reset_index()
    .melt(id_vars = ['country', 'continent'],
         value_name = 'lifeExpectacy')
)
```

## Subset data for later use

- Read in 'gapminder_data.csv' data
- filter out year 2007 and continent Americas
- drop the year and continent columns
- save the new dataframe into a variable called gapminder_2007

```python
# - Read in 'gapminder_data.csv' data
gapminder = pd.read_csv('./data/gapminder_data.csv')
```

```python
# - filter out year 2007 and continent Americas
# - drop the year and continent columns
# - save the new dataframe into a variable called gapminder_2007
gapminder_2007 = (
    gapminder
    .query("year == 2007")
    .query("continent == 'Americas'")
    .drop(columns = ['year', 'continent'])
)
```
